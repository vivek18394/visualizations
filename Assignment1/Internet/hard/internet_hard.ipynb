```python
# Libraries
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd

# Make data: I have 10 groups and some subgroups
share_df = pd.read_csv('share-of-individuals-using-the-internet.csv')

#Entity, code, year, percentage

share_df.shape #6056 X 4

#create a dataframe which has all countries and percentage in year 2016

share_2016_df = share_df[share_df['Year'] == 2016]

share_2016_df = share_2016_df[["Entity", "Individuals using the Internet (% of population)"]]

region = ["Pacific island small states", "South Asia"] #to be printed in outer donut

boolean_series = share_2016_df.Entity.isin(region)

share_2016_regions_df = share_2016_df[boolean_series]

group_size = share_2016_regions_df["Individuals using the Internet (% of population)"] #total number of internet users in south asia and pacific islands

print(group_size)

subregion_names = ['Bhutan', 'Maldives','Micronesia (country)', 'Fiji', 'French Polynesia',
                'Kiribati','Marshall Islands','New Zealand',
                'Solomon Islands','Tonga','Tuvalu','Vanuatu', 'Afghanistan', 
                   'India', 'Pakistan', 'Sri Lanka', 'Nepal']#countries in s asia and PI

subregion_series = share_2016_df.Entity.isin(subregion_names)

share_2016_subregions_df = share_2016_df[subregion_series]#size of each country in s.asia and PI

subgroup_size = share_2016_subregions_df["Individuals using the Internet (% of population)"]
sub_size = list(subgroup_size)

print(share_2016_subregions_df)

# Create colors
a, b=[plt.cm.Blues, plt.cm.Reds]

# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie([320,107.37], radius=3.3, labels=region, colors=[a(0.8), b(0.90)] )
plt.setp( mypie, width=1.4, edgecolor='white')


# Second Ring (Inside)
sub_colors = [a(0.7),a(0.65),a(0.60), a(0.55),a(0.50),a(0.45),a(0.40), a(0.34),a(0.28),a(0.22),
              a(0.18),a(0.12), b(0.80) ,b(0.72),b(0.63), b(0.57),b(0.50)] 
print(type(sub_colors))
mypie2, _ = ax.pie(sub_size, radius=3.3-1.3, labels=subregion_names, labeldistance=1.1,
                   colors=sub_colors,textprops={'color':"w",'fontsize': 10,'weight':'bold'})

plt.setp( mypie2, width=1.0, edgecolor='white')
plt.margins(0,0)
 
# show it
plt.show()
```
