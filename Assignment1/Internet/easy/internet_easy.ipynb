```python
#Easy: Bar chart

#visualize the data for these regions as a pie chart and as a bar plot - using the 
#latest year for which data is available for all these regions (i.e. it should be the same year for all regions)

import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd

share_df = pd.read_csv('share-of-individuals-using-the-internet.csv')
#Entity, code, year, percentage

share_df.shape #6056 X 4

#create a dataframe which has all countries and percentage in year 2016

share_2016_df = share_df[share_df['Year'] == 2016]

#filter dataframe with just two columns: Entity and % of interent users.

share_2016_df = share_2016_df[["Entity", "Individuals using the Internet (% of population)"]] 

#filter dataframe with following regions: Arab World, Caribbean small states, Central Europe and the Baltics, 
#East Asia & Pacific, Europe & Central Asia, Latin America & Caribbean, 
#Middle East & North Africa, Pacific island small states, South Asia, Sub-Saharan Africa.

region = ["Arab World", "Caribbean small states", "Central Europe and the Baltics", 
"East Asia & Pacific", "Europe & Central Asia", "Latin America & Caribbean", 
"Middle East & North Africa", "Pacific island small states", "South Asia", "Sub-Saharan Africa"]


boolean_series = share_2016_df.Entity.isin(region)

share_regions_2016_df = share_2016_df[boolean_series]

share_regions_2016_df = share_regions_2016_df.sort_values(by=['Individuals using the Internet (% of population)'],ascending=True)

share_regions_2016_df.plot.bar(x="Entity", y="Individuals using the Internet (% of population)", rot=70, title="Individuals using the Internet (% of population)");
plt.xlabel("Regions")
plt.ylabel("(% of population)")

plt.show(block=True);
```

```python
#Easy: Pie chart
share_regions_2016_df.plot.pie(y = "Individuals using the Internet (% of population)",
                               labels = share_regions_2016_df["Entity"],figsize=(11, 11), legend=None)
plt.axes().set_ylabel('')
plt.show(block=True)
```
